Import Library 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

Import Data

```python
data = pd.read_csv('dataset_penjualan_hp.csv')
data
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.rename (
    columns = {'Tanggal Transaksi' : 'Tanggal_Transaksi', 
               'Lokasi Pembelian' : 'Lokasi_Pembelian', 
               'Jenis Kelamin' : 'Jenis_Kelamin', 
               'Jenis Pembayaran' : 'Jenis_Pembayaran', 
               'Harga Jual' : 'Harga_Jual'})
data
```

```python
data_lokasi = data.Lokasi_Pembelian.value_counts()
ax= sns.barplot(data_lokasi.index, data_lokasi.values, alpha = 0.8)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha= 'right')
plt.title('Total Penjualan Hanphone Berdasarkan Lokasi Pembelian')
plt.xlabel('Lokasi')
plt.ylabel('')
```

```python
data_Kota = data.Kota.value_counts()
ax= sns.barplot(data_Kota.index, data_Kota.values, alpha = 0.8)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha= 'right')
plt.title('Total Penjualan Hanphone Berdasarkan Kotas Asal Pembeli')
plt.xlabel('Kota Asal')
plt.ylabel('')
```

```python
data_tipe = data.Tipe.value_counts()
data_tipe = data_tipe[0:10,]
ax= sns.barplot(data_tipe.values, data_tipe.index, alpha = 0.8)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha= 'right')
plt.title('Top 10 Job Handphone yang Paling Diminati')
plt.xlabel('')
plt.ylabel('Tipe Handphone')
```

```python
data_merek = data.Merek.value_counts()
ax= sns.barplot(data_merek.index, data_merek.values, alpha = 0.8)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha= 'right')
plt.title('Total Penjualan Hanphone Berdasarkan Merek')
plt.xlabel('Merek Handphone')
plt.ylabel('')
```

```python
data_warna = data.Warna.value_counts()
ax= sns.barplot(data_warna.index, data_warna.values, alpha = 0.8)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha= 'right')
plt.title('Total Penjualan Hanphone Berdasarkan Warna')
plt.xlabel('Lokasi')
plt.ylabel('')
```

```python
data_gender = data.Jenis_Kelamin.value_counts()
ax= sns.barplot(data_gender.index, data_gender.values, alpha = 0.8)
ax.set_xticklabels(ax.get_xticklabels(), ha= 'center')
plt.title('Total Penjualan Hanphone Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('')
```

```python
sns.set(style="ticks", palette="pastel")
sns.boxplot(x= "Merek", y = "Usia", hue = "Jenis_Kelamin", palette=["m", "g"],
            data=data)
```

```python
sns.set(style = "whitegrid")
sns.boxplot(y="Harga_Jual", data = data)
```
